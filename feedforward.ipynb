```python
D = 2 # dimensions
M = 3 # Hidden layers
K = 3 # output classes
```

```python
w1 = np.random.randn(D,M) #weight for hidden layer
b1 = np.random.randn(M) # bias for hidden layer
w2 = np.random.randn(M,K) #weight for output layer
b2 = np.random.randn(K) # bias for output layer
```

```python
print(w1)
print(b1)
print(w2)
print(b2)
```

```python
def sigmoid(a):
    return 1/(1+np.exp(-a))
```

```python
def forward(X,w1,b1,w2,b2):
    c = X.dot(w1)+b1
    #print(c)
    Z = sigmoid(c) #sigmoid
    #print(Z)
    Z1 = np.tanh(c) #tanh
    Z2 = np.maximum(c,0) # ReLu
    A = Z.dot(w2)+b2 #output layer
    expA = np.exp(A) #exponential values for A
    Y = expA/expA.sum(axis = 1,keepdims = True) #probabilities for all values to get value 1(softmax) after adding every row values
    #print(Y.shape)
    return Y # softmax values
```

```python
forward(X,w1,b1,w2,b2)
```

```python
#classification rate
def classification_rate(Y,P):
    n_correct = 0
    n_total = 0
    for i in range(len(Y)):
        n_total += 1
        if Y[i] == P[i]:
            n_correct += 1
    return float(n_correct)/n_total        
```

```python
P_Y_X = forward(X,w1,b1,w2,b2)
```
