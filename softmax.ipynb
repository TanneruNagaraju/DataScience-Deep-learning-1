```python
import numpy as np
```

```python
a = np.random.randn(5)
```

```python
a
```

```python
expa = np.exp(a) # exponential values for value A
```

```python
expa
```

```python
output = expa/expa.sum() # softmax formula 
output1 = expa/np.sum(expa) # another type
```

```python
output1
```

```python
output #probabilities for softmax values
```

```python
output.sum() # adding all probabilities to get value 1 
```

# Another Example


```python
A = np.random.randn(100,5)
```

```python
A
```

```python
expA = np.exp(A) #exponential values for A
```

```python
expA
```

```python
answer = expA/np.sum(expA)
```

```python
answer
```

```python
answer.sum()
```

```python
ans = expA/expA.sum(axis = 1,keepdims = True)
```

```python
ans
```

```python
ans.sum(axis = 1) #all column values are adding to get value 1
```
